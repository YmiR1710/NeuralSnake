```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/Getty keywords and metadata tags.xlsx')
df
```

```python
df['combined'] = df['LOCATION'] + ',' + df['PEOPLE_ETHNICITY'] + ',' + df['PEOPLE_GENDER'] + ',' + df['PEOPLE_NUMBER'] + ',' + df['PEOPLE_ROLE'] + ',' + df['EVENT'] + ',' + df['IMAGE_INDUSTRY'] + ',' + df['LOB'] + ',' + df['OBJECTS'] + ',' + df['ORIENTATION'] + ',' + df['REGIONAL'] + ',' + df['RESTRICTED'] + ',' + df['THEMES'] + ',' + df['PEOPLE_AGE']
def clean_and_combine(row):
    columns = ['KEYWORDS', 'LOCATION', 'PEOPLE_ETHNICITY', 'PEOPLE_GENDER', 'PEOPLE_NUMBER', 'PEOPLE_ROLE', 'EVENT', 'IMAGE_INDUSTRY', 'LOB', 'OBJECTS', 'ORIENTATION', 'REGIONAL', 'RESTRICTED', 'THEMES', 'PEOPLE_AGE']
    combined = ','.join([str(row[col]).lower() for col in columns if pd.notnull(row[col])])
    return combined

df['combined'] = df.apply(clean_and_combine, axis=1)
df['combined'][0]
```

```python
from itertools import product
from collections import Counter


def generate_keyword_tag_pairs(row):
    # Split keywords into a list, assuming 'keywords' column exists and avoiding NaN
    keywords = [kw.strip() for kw in str(row['KEYWORDS']).split(',') if pd.notnull(row['KEYWORDS'])]

    # Initialize an empty list to store the keyword-tag pairs for this row
    pairs = []

    # List all tag columns
    tag_columns = ['LOCATION', 'PEOPLE_ETHNICITY', 'PEOPLE_GENDER', 'PEOPLE_NUMBER', 'PEOPLE_ROLE', 'EVENT', 'IMAGE_INDUSTRY', 'LOB', 'OBJECTS', 'ORIENTATION', 'REGIONAL', 'RESTRICTED', 'THEMES', 'PEOPLE_AGE']

    for tag_column in tag_columns:
        # Continue only if the tag column is not NaN
        if pd.notnull(row[tag_column]):
            # Split this tag column's value into separate tags and strip whitespace
            tags = [tag.strip() for tag in str(row[tag_column]).split(',')]
            # For each tag, pair with each keyword
            for tag in tags:
                for keyword in keywords:
                    # Ensure tag is not being paired with itself as keyword
                    if keyword != tag:
                        pairs.append((tag, keyword))

    return pairs


pair_counts = Counter()

for index, row in df.iterrows():
    pairs = generate_keyword_tag_pairs(row)
    pair_counts.update(pairs)

sorted_pairs = sorted(pair_counts.items(), key=lambda x: x[1], reverse=True)
```

```python
# Assume sorted_pairs is your sorted list of (tag, keyword) pairs with counts
sorted_pairs_df = pd.DataFrame(sorted_pairs, columns=['Tag-Keyword Pair', 'Frequency'])

# Display the first few rows to verify
sorted_pairs_df
```
